```python
X = np.random.random(size=(1000, 2)) * 10
y = 5 + (np.array([2, 6]) * X).sum(axis=1) + np.random.normal(loc=0, scale=0.1)

# add bias term
X = np.hstack([np.ones((1000,1)), X])
(X.shape, y.shape)
```

# closed form solution

```python
# w = (X'X)^-1 * X'y
```

```python
w = np.linalg.inv(X.T @ X) @ X.T @ y
```

# gradient descent

```python
alpha = 0.01
```

```python
w = np.random.random((3,))
w
```

```python
for idx in range(10000):
    gradient = X.T @ (X @ w - y)
    if idx % 1000 == 0:
        print(np.abs(gradient).sum())
        # print(gradient)
    w = w - (alpha / X.shape[0]) * gradient
```
